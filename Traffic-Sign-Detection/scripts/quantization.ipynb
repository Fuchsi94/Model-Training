# Dynamic range quantization

#Full Integer Quantization

```python
def representative_data_gen():
  dataset_list = tf.data.Dataset.list_files('/content/Traffic-Sign-Data/train/*.jpeg')
  for i in range(100):
    image = next(iter(dataset_list))
    image = tf.io.read_file(image)
    image = tf.io.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, [320, 320])
    image = tf.cast(image / 255., tf.float32)
    image = tf.expand_dims(image, 0)
    yield [image]
```

```python
import tensorflow as tf
import numpy as np
tfVersion=tf.version.VERSION.replace(".", "")# can be used as savename
print("tensorflow ", tf.version.VERSION)


#CONVERT YOUR MODEL TO TFLITE AND QUANTIZE TO INT8 
converter = tf.lite.TFLiteConverter.from_saved_model('/content/saved_model')

converter.optimizations = [tf.lite.Optimize.DEFAULT]
converter.representative_dataset = representative_data_gen
converter.target_spec.supported_ops = [tf.lite.OpsSet.TFLITE_BUILTINS_INT8]
converter.experimental_new_converter = True

converter.target_spec.supported_types = [tf.int8]
converter.inference_input_type = tf.int8 
converter.inference_output_type = tf.int8 
quantized_tflite_model = converter.convert()

#SAVE THE QUANTIZED MODEL AS TFLITE FILE
open('/content/saved_model/quant_model.tflite' , "wb").write(quantized_tflite_model)
```
